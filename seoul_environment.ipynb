```python
import pandas as pd
import streamlit as st

df1 = pd.read_csv('./data/seoul_2008_2011.csv', encoding='cp949')
df2 = pd.read_csv('./data/seoul_2012_2015.csv', encoding='cp949')
df3 = pd.read_csv('./data/seoul_2016_2019.csv', encoding='cp949')
df4 = pd.read_csv('./data/seoul_2020_2021.csv', encoding='cp949')
df5 = pd.read_csv('./data/seoul_2022.csv', encoding='cp949')
```

```python
#df5 typo in col micro fine dust
df1.head()
```

```python
df = pd.concat([df1,df2,df3,df4])
df
#3405313 rows × 5 columns
```

```python
#change col names
df.columns =['Date', 'Neighborhood', 'Finedust(PM10)', 'Microdust(PM2.5)']
df.head
```

```python
#change col names of df5 and concate
df5.columns =['Date', 'Neighborhood', 'Finedust(PM10)', 'Microdust(PM2.5)']
df = pd.concat([df, df5])
df
#3405313 rows × 4 columns
```

```python
#시간 추출
df['Hour'] = df['Date'].apply(lambda x: x[11:13])
df['Date'] = df['Date'].apply(lambda x: x[:11])
df = df[['Date','Hour','Neighborhood','Finedust(PM10)','Microdust(PM2.5)']]
df.head()
```

```python
#separate by year and month
df['Year'] = df['Date'].apply(lambda x: x[:4])
df['Month'] = df['Date'].apply(lambda x: x[5:7])
df['Days'] = df['Date'].apply(lambda x: x[8:10])
df.head()
```

```python
#sort by year and month
df = df.sort_values(by = ['Date'], ascending= True)
df
```

```python
df.groupby('Neighborhood').mean()['Finedust(PM10)']
```

```python
df.to_csv('./data/seoul_dust_added.csv', encoding='cp949', index =False)
```

```python
#pivot table 
df_by_neighborhood = df.pivot_table(index="Date", 
                              columns="Neighborhood", 
                              values="Microdust(PM2.5)")
df_by_neighborhood
```
